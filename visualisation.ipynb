```python
#Bar chart to display the top 10 car brand with the most sale

agg = df.groupby("car").size().sort_values(ascending=False)
top10 = agg.head(10)
x=top10.index
y=top10.values

plt.figure(figsize=(10,6))
plt.bar(x,y, color=COL )
plt.ylabel("Amount Sold")
plt.xlabel("Brand name")
plt.xticks(rotation=45, ha="right")
plt.title("Most sold cars (Top 10)")
plt.ylim(0, max(y)*1.1)
for h,v in enumerate(y):
    plt.text(h,v,str(v),ha="center", va="bottom")
    
plt.show()
```

```python
#Display top 10 years of highest car manufacturing

aggYear = df.groupby("year").size().sort_values(ascending=False)
top10Year = aggYear.head(10)
x=top10Year.index.astype(str)
y=top10Year.values

plt.figure(figsize=(10,6))
plt.plot(x, y,  color=COL )
plt.fill_between(x,y,color=COL, alpha=0.3)
plt.ylabel("Amount Manufactured")
plt.xlabel("Year")
plt.xticks(rotation=45, ha="right")
plt.title("Peak manufacturing years (Top 10)")
plt.ylim(0, max(y)*1.1)

for h,v in enumerate(y):
    plt.text(h,v,str(v),ha="center", va="bottom")
    
plt.show()
```


# Correlation Heatmap of Car Features Explanation

This image shows a **correlation heatmap** of four car features: `price`, `mileage`, `engv` (engine volume), and `year`. The heatmap visualizes the correlation coefficients between these features with a color gradient and numerical annotations.

## Visual Details
- **Title**: "Correlation Heatmap of Car Features".
- **Axes**: The x-axis and y-axis list the features: `price`, `mileage`, `engv`, and `year`.
- **Color Scale**: 
  - Dark green : Strong positive correlation (1.0).
  - medium green: Weak or no correlation (near 0).
  - light green: Negative correlation (e.g., -0.4).
- **Annotations**: Each cell displays the correlation coefficient between the two features at the corresponding row and column.

## Correlation Insights 
- **Price vs. Price, Mileage vs. Mileage, etc.**: 1.0 (dark green, perfect positive correlation with self).
- **Price vs. Mileage**: -0.31 (moderate negative correlation, price decreases as mileage increases).
- **Price vs. Year**: 0.38 ( moderate positive correlation, newer cars cost more).
- **Mileage vs. Year**: 0.49 ( moderate positive correlation, newer cars have lower mileage).
- **Price vs. engv**: 0.048 ( very weak positive correlation).
- **Mileage vs. engv**: 0.047 ( very weak positive correlation).
- **engv vs. Year**: -0.042 ( very weak negative correlation).

The heatmap highlights strong self-correlations (1.0) on the diagonal and varying relationships between features, with moderate correlations indicating meaningful patterns and weak correlations showing little to no linear relationship.

# Price Distribution by Body Type <span style="color: #6af2a3;">📊</span>

## Overview
This boxplot illustrates the distribution of car prices across different vehicle body types. Prices are shown on the y-axis, and body types are categorized on the x-axis.

## Data Representation
- *X-Axis (Body Types)*: 
  - Crossover
  - Sedan
  - Other body
  - Van
  - Wagon
  - Hatch
  <span style="color: #6af2a3;">(Categories are evenly spaced along the horizontal axis)</span>

- *Y-Axis (Price)*: 
  - Range: 0 to 500,000
  <span style="color: #6af2a3;">(Vertical axis in monetary units, likely USD)</span>

## Boxplot Details
Each boxplot represents the price distribution for a specific body type, with the following components:
- *Box*: Shows the interquartile range (IQR) between the 25th (Q1) and 75th (Q3) percentiles.
- *Median Line*: Central line within each box, indicating the 50th percentile.
- *Whiskers*: Extend to the minimum and maximum values within 1.5 times the IQR from Q1 and Q3.
- *Outliers*: Dots beyond the whiskers, indicating extreme prices.
